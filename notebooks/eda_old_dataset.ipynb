```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
df = pd.read_csv(Path.cwd().parent.parent / "credit_risk_dataset.csv")
df.head()
```

```python
df.info()
```

# Univariate Analysis

```python
df["person_age"].hist(bins=30)

plt.xlabel("Interest Rate")
plt.ylabel("Frequency")
plt.title("Distribution of Age")
plt.show()
```

```python
df["loan_percent_income"].hist(bins=100)

plt.xlabel("Income")
plt.ylabel("Frequency")
plt.title("Distribution of Income")
plt.show()
```

```python
df['person_home_ownership'].value_counts(ascending=True).plot(kind='barh')

plt.xlabel('Count')
plt.ylabel('Loan Intent')
plt.title('Distribution of Home Ownership')
plt.show()
```

```python
df["person_emp_length"].hist(bins=100)

plt.xlabel("Employment Length")
plt.ylabel("Frequency")
plt.title("Distribution of Employment Length")
plt.show()
```

```python
df['loan_intent'].value_counts(ascending=True).plot(kind='barh')

plt.xlabel('Count')
plt.ylabel('Loan Intent')
plt.title('Distribution of Loan Intent')
plt.show()
```

```python
df['loan_grade'].value_counts(ascending=True).plot(kind='barh')

plt.xlabel('Count')
plt.ylabel('Loan Grade')
plt.title('Distribution of Loan Grade')
plt.show()
```

```python
df["loan_amnt"].hist(bins=100)

plt.xlabel("Income")
plt.ylabel("Frequency")
plt.title("Distribution of Loan Amounts")
plt.show()
```

```python
df["loan_int_rate"].hist(bins=100)

plt.xlabel("Interest Rate")
plt.ylabel("Frequency")
plt.title("Distribution of Interest Rates")
plt.show()
```

```python
df['loan_status'].value_counts(ascending=True).plot(kind='barh')

# 0 is non default, 1 is default

plt.xlabel('Count')
plt.ylabel('Loan Status')
plt.title('Distribution of Loan Status')
plt.show()
```

```python
df["loan_percent_income"].hist(bins=100)

plt.xlabel("Percent Income")
plt.ylabel("Frequency")
plt.title("Distribution of Loan as Percent of Income")
plt.show()
```

```python
df['cb_person_default_on_file'].value_counts(ascending=True).plot(kind='barh')

# historical default

plt.xlabel('Count')
plt.ylabel('Historical Default')
plt.title('Distribution of Historical Default')
plt.show()
```

```python
df["cb_person_cred_hist_length"].hist(bins=30)

plt.xlabel("Credit History Length")
plt.ylabel("Frequency")
plt.title("Distribution of Loan as Percent of Income")
plt.show()
```
